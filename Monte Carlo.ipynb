# Monte Carlo Integration

Monte Carlo Integration is a numerical method that uses **random sampling** to approximate definite integrals, especially useful for high-dimensional problems.

---

## Theory

For a one-dimensional integral

$$
I = \int_a^b f(x) \, dx
$$

we can approximate it by sampling $N$ random points $x_i$ uniformly in $[a, b]$:

$$
I \approx \frac{b-a}{N} \sum_{i=1}^N f(x_i)
$$

- $x_i \sim \mathcal{U}(a, b)$ are uniformly distributed random numbers.  
- As $N \to \infty$, the estimate converges to the true integral.  
- The error decreases roughly as $\mathcal{O}(1/\sqrt{N})$.

---

## Example: Estimating $\pi$

The value of $\pi$ can be estimated by sampling random points $(x, y)$ in the unit square $[0,1] \times [0,1]$.  
The fraction of points inside the quarter circle of radius 1 gives:

$$
\pi \approx 4 \cdot \frac{\text{Number of points inside circle}}{\text{Total number of points}}
$$

---


```python
import numpy as np
import matplotlib.pyplot as plt
def value_of_pi(total_points):
    #total_points = 100
    inside_the_circle = 0
    for i in range(total_points+1):
        x, y = np.random.random(), np.random.random()
        if y < np.sqrt(1.0-x*x):
            inside_the_circle += 1
    return 4.0* (inside_the_circle/total_points)

n = list(range(10,10000,10))
pi = []
for i in n:
    pi.append(value_of_pi(i))
dim = len(pi)
exact_value = [np.pi]*dim
plt.plot(n,pi,"b-",label="Approximate value")
plt.plot(n,exact_value,"r--",label="Exact value")
plt.xlim(0,max(n))
plt.legend()
plt.show()
```
